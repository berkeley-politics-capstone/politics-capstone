```python
import pandas as pd
import os
from glob import iglob
import json

DATA_DIRECTORY = "../../data/Youtube/Lists_of_Videos/"
```

```python
rootdir_glob = DATA_DIRECTORY + '**/*' # Note the added asterisks
# This will return absolute paths
file_list = [f for f in iglob(rootdir_glob, recursive=True) if os.path.isfile(f)]
```

```python
data = []
for file in file_list:
    with open(file, 'r') as f:
        j = json.load(f)
        for i in range(50):
            data.append((j["items"][i]['id']['videoId']
                        ,file.split("/")[5]
                        ,file.split("/")[6]
                        ,i+1
                        ))
df = pd.DataFrame(data, columns=('videoId', 'candidate', 'day_observed', 'rank'))
df
```

```python
for candidate in set(df.candidate):
    print(candidate, df[df.candidate == candidate][['candidate', 'channelTitle', 'videoId']]\
                        .groupby(('candidate','channelTitle'))\
                        .agg('count')\
                        .sort_values('videoId', ascending=False)\
                        .head(10))
```
