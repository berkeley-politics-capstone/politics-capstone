```python
import pandas as pd
from multiprocessing import Pool, Manager, Value, Lock

from tqdm._tqdm_notebook import tqdm_notebook as tqdm
tqdm.pandas()

from newsplease import NewsPlease
from itertools import product
import os
import pickle

from datetime import datetime

import warnings
warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')
```

```python

DATA_DIR = "../../data/reddit/Article_data_2019/"
DATA_FILE = "reddit_2019jun16tojul1_articleurls"
data = pd.read_pickle(DATA_DIR + DATA_FILE + ".pkl")
```

```python
def get_data(url):
    with lock:
        counter.value += 1
    if counter.value >= 1:
        os.write(1, "{}\n".format(counter.value).encode())
    if url not in web_data:
        try:
            if url is not None and not url.endswith('.pdf'):
                news = NewsPlease.from_url(url, timeout=5)
                web_data[url] = news.get_serializable_dict()
        except Exception as e:
            os.write(1, url.encode())
            os.write(1, str(datetime.now()).encode())
```
