```python
import zipfile
from newsplease import NewsPlease
from tqdm import tqdm_notebook as tqdm

with zipfile.ZipFile('articlelinks.zip', 'r').open("articlelinks.txt", "r") as f:
    data = eval(f.readline())
    
missed_rows = []
try:
    with open("article_data.csv", 'w') as f:
        for i in tqdm(range(len(data))):
            try:
                news = NewsPlease.from_url(data[i], timeout=5)
                f.write("{}, {} \n".format(data[i], news.get_serializable_dict()))
            except Exception as e:
                missed_rows.append((i, data[i]))
except Exception as f:
    print(missed_rows)
```
