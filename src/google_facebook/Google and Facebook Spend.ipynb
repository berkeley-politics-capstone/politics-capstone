# Google and Facebook Spend

Both [Google](https://transparencyreport.google.com/political-ads/region/US) and [Facebook](https://www.facebook.com/ads/library/report/) publish reports on political advertising for the goal of increasing transparency. 

[Bully Pulpit Media](https://bpimedia.com/) uses this data to publish a weekly metric report called the [2020 Campaign Tracker](https://2020campaigntracker.com/). This site reports each candidate's weekly spend by both Facebook and Google platforms.

I've downloaded a CSV file of the 2020 Campaign Tracker, which has already cleaned up the data before visualizing it (in particular, they map the fund/PAC name that is used in Google and Facebook to the candidate). This file was last updated **May 30, 2019.**

```python
# Libraries

import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
# Bringing in the CSV file

df = pd.read_csv('2020_Digital_Spend.csv') 
```

```python
df
```

```python
# Reproducing the chart that 2020 Campaign Tracker does

df.groupby(['Candidate', 'Platform'])['Spend'].sum().unstack('Platform').fillna(0).plot(kind='bar', stacked=True)
```
